```
import numpy
from matplotlib import pyplot
%matplotlib inline
import seaborn as sns
```

```
x = numpy.arange(1.0, 1000, 10.0)

fig, ax = pyplot.subplots(3,3, figsize=(16,12))
sns.set_style('darkgrid')
for i, txt in enumerate(numpy.arange(-0.4, 0.5, 0.1)):
  y = numpy.exp(x**txt)
  j = i//3
  k = i%3
  sns.scatterplot(x = x, y = y, ax=ax[j][k], palette = 'rainbow')
  ax[j][k].set_title(str(i+1) + ' Exponential of x to the power of ' + str(txt))
  ax[j][k].set_xlabel('x')
  ax[j][k].set_ylabel('y = exp (x^'+str(txt)+')')

fig.tight_layout()
pyplot.show()
print('All done!')
```
